### Data Cleaning and EDA

```python
cat = ['Description','Street','City','Zipcode','Timezone','Airport_Code','Weather_Timestamp','Wind_Direction','Weather_Condition','Sunrise_Sunset','Civil_Twilight','Nautical_Twilight','Astronomical_Twilight']
```

```python
for i in cat:
    mean = df[i].mode()[0]
    df[i].fillna(mean,inplace = True)
```

```python
continues = ['End_Lat','End_Lng','Temperature(F)','Wind_Chill(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Wind_Speed(mph)','Precipitation(in)']
```

```python
for i in continues:
    mean1= df[i].mean()
    df[i].fillna(mean1,inplace = True)
```

```python
# After cleaning the data
df.isnull().sum()
```

```python
# Analyze patterns related to road conditions, weather, and time of day
# For example, let's check the distribution of road conditions
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Weather_Condition', order=df['Weather_Condition'].value_counts().index)
plt.title('Distribution of Weather Conditions')
plt.xlabel('Count')
plt.ylabel('Weather Condition')
plt.show()
```

```python
# Visualize accident hotspots (latitude and longitude)
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df.sample(1000), x='Start_Lng', y='Start_Lat', hue='Severity', palette='viridis', alpha=0.7)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Severity')
plt.show()
```

```python
# Visualize contributing factors
# For example, let's check the distribution of contributing factors
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Traffic_Signal', order=df['Traffic_Signal'].value_counts().index)
plt.title('Distribution of Traffic Signal as a Contributing Factor')
plt.xlabel('Count')
plt.ylabel('Traffic Signal')
plt.show()
```

```python
# Convert Start_Time to datetime for time-based analysis
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

# Extract date, hour, and day of the week from Start_Time
df['Date'] = df['Start_Time'].dt.date
df['Hour'] = df['Start_Time'].dt.hour
df['Day_of_Week'] = df['Start_Time'].dt.day_name()
```

```python
# Analyze distribution of accidents by day of the week
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Day_of_Week', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Accident Distribution by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyze distribution of accidents by hour of the day
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Hour')
plt.title('Accident Distribution by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Correlation Analysis:

```python
# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

### To explore factors influencing accident severity, we'll analyze how location, weather, and time of day relate to accident severity.

```python
# Plotting accident severity distribution by state
plt.figure(figsize=(24, 8))
sns.countplot(data=df, y='State', hue='Severity', hue_order=[1, 2, 3, 4], dodge=False)
plt.title('Accident Severity by State')
plt.xlabel('Count')
plt.ylabel('State')
plt.legend(title='Severity')
plt.show()
```

```python
# Plotting accident severity distribution by weather condition
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Weather_Condition', hue='Severity', hue_order=[1, 2, 3, 4], dodge=False)
plt.title('Accident Severity by Weather Condition')
plt.xlabel('Count')
plt.ylabel('Weather Condition')
plt.legend(title='Severity')
plt.show()
```

```python
# Plotting accident severity distribution by time of day
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Hour', hue='Severity', hue_order=[1, 2, 3, 4], dodge=False)
plt.title('Accident Severity by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Count')
plt.legend(title='Severity')
plt.show()
```

### Hourly Trend Analysis:

For time series analysis, we'll analyze trends and patterns in accidents over time, focusing on temporal aspects such as hour of the day and month of the year

```python
# Grouping by hour and counting accidents
hourly_accidents = df.groupby('Hour').size()

# Plotting the hourly trend of accidents
plt.figure(figsize=(12, 6))
hourly_accidents.plot(kind='line', marker='o')
plt.title('Hourly Trend of Accidents')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.xticks(range(24))
plt.grid()
plt.show()
```

### Multivariate Analysis using Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting relevant columns for multivariate analysis
selected_columns = ['Weather_Condition', 'Visibility(mi)', 'Severity']

# Creating a subset of the dataframe with selected columns
subset_df = df[selected_columns]

# Dropping rows with any missing values in the selected columns
subset_df.dropna(subset=selected_columns, inplace=True)

# Creating a heatmap to explore relationships
plt.figure(figsize=(10, 8))
heatmap_data = pd.pivot_table(subset_df, index='Weather_Condition', columns=['Visibility(mi)'], 
                               values='Severity', aggfunc='mean')
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt='.2f', cbar=True, square=True)
plt.title('Relationships between Weather, Visibility, and Severity')
plt.xlabel('Visibility (mi)')
plt.ylabel('Weather Condition')
plt.show()
```
